```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

```python
data_2d = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data_2d.csv', header=None)
mnist = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/mnist.csv', header=None)
```

```python
X_2d = data_2d.values
X_mnist = mnist.values

scaler_2d = StandardScaler()
scaler_mnist = StandardScaler()
X_2d_scaled = scaler_2d.fit_transform(X_2d)
X_mnist_scaled = scaler_mnist.fit_transform(X_mnist)
```

## data_2d.csv

```python
k_values_2d = range(2, 11)
inertias_2d = []

for k in k_values_2d:
    kmeans = KMeans(n_clusters=k, random_state=20, n_init="auto")
    kmeans.fit(X_2d_scaled)
    inertias_2d.append(kmeans.inertia_)

plt.plot(k_values_2d, inertias_2d, 'bo-')
plt.xlabel('Cluster count')
plt.ylabel('Distortion')
plt.title('data_2d.csv')
plt.show()
```

```python
optimal_k_2d = 5
kmeans_2d = KMeans(n_clusters=optimal_k_2d, random_state=20, n_init="auto")
kmeans_2d.fit(X_2d_scaled)
labels_2d = kmeans_2d.labels_
```

```python
print("Результати кластеризації (data_2d.csv):")
print(labels_2d)
```

##mnist.csv

```python
k_values_mnist = range(2, 11)
inertias_mnist = []

for k in k_values_mnist:
    kmeans = KMeans(n_clusters=k, n_init="auto", random_state=30)
    kmeans.fit(X_mnist_scaled)
    inertias_mnist.append(kmeans.inertia_)

plt.plot(k_values_mnist, inertias_mnist, 'bo-')
plt.xlabel('Cluster count')
plt.ylabel('Distortion')
plt.title('mnist.csv')
plt.show()
```

```python
optimal_k_mnist = 4
kmeans_mnist = KMeans(n_clusters=optimal_k_mnist, random_state=32, n_init="auto")
kmeans_mnist.fit(X_mnist_scaled)
labels_mnist = kmeans_mnist.labels_
```

```python
print("Результати кластеризації (mnist.csv):")
print(labels_mnist)
```

##Visualation

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
X_mnist_pca = pca.fit_transform(X_mnist_scaled)

plt.scatter(X_2d[:, 0], X_2d[:, 1], c=labels_2d)
plt.title('Кластеризація (data_2d.csv)')
plt.show()

plt.scatter(X_mnist_pca[:, 0], X_mnist_pca[:, 1], c=labels_mnist)
plt.title('Кластеризація (mnist.csv)')
plt.show()
```
